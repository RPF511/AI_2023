```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_seeds = np.array([(0,0),(0,1),(1,0),(1,1)], dtype = np.float64)
y_seeds = np.array([0,1,1,0])

# N = 1000
# idxs = np.random.randint(0,4,N)

# X = x_seeds[idxs]
# Y = y_seeds[idxs]
# print(X)
# X += np.random.normal(scale = 0.25, size = X.shape)
# print(X)
```

```python
class ShallowNN:
    def __init__(self,seed_x,seed_y, num_input_features, num_hiddens, lr = 0.25):
        self.seed_X = seed_x
        self.seed_Y = seed_y
        self.w1 = np.random.normal(size = (num_hiddens, num_input_features))
        self.b1 = np.random.normal(size = num_hiddens)
        self.w2 = np.random.normal(size = (num_hiddens))
        self.b2 = np.random.normal(size = 1)
        self.lr = lr
        self.data_X = np.array([])
        self.data_Y = np.array([])
        self.history_loss = [[],[]]
        return
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def mk_train_dataset(self, N = 1000):
        idxs = np.random.randint(0,4,N)

        self.data_X = self.seed_X[idxs]
        self.data_Y = self.seed_Y[idxs]
#         print(X)
        self.data_X += np.random.normal(scale = 0.25, size = self.data_X.shape)
#         print(X)
        return
    
    def predict(self, x):
        if type(x) != np.ndarray:
            x = np.array(x)
        z1 = np.matmul(self.w1, x) + self.b1
#         print(z1)
        a1 = np.tanh(z1)
#         print(a1)
        z2 = np.matmul(self.w2, a1) + self.b2
#         print(z2)
        a2 = self.sigmoid(z2)
        return a2, (z1,a1,z2,a2)
    
    def train(self, epoch_num = 0, lr = 0):
        if(not lr):
            lr = self.lr
        if(epoch_num):
            print("\n[epoch ",epoch_num,"]\n    {before} w1 ",self.w1," , b1 ",self.b1," , w2 ",self.w2," , b2 ",self.b2,)             
                  
        dw1 = np.zeros_like(self.w1)
        db1 = np.zeros_like(self.b1)
        dw2 = np.zeros_like(self.w2)
        db2 = np.zeros_like(self.b2)
        
        m = len(self.data_X)
        cost = 0.0
        
        for x,y in zip(self.data_X, self.data_Y):
            a2, (z1,a1,z2,_) = self.predict(x)
            if(y):
                cost -= np.log(a2)
            else:
                cost -= np.log(1-a2)
                
            diff = a2-y
            db2 += diff
            
            dw2 += a1 * diff
#             print(dw2)
            db1_diff = (1 - a1**2)*self.w2 * diff
            db1 += db1_diff
#             print(db1)
            dw1 += np.matmul(np.array([db1_diff]).T , np.array([x]))
#             print(np.array([db1_diff]).T)
#             print(np.array([x]))
#             print()
#             print(dw1)
#             print()
        cost /= m
        self.w1 -= lr * dw1/m
        self.b1 -= lr * db1/m
        self.w2 -= lr * dw2/m
        self.b2 -= lr * db2/m
        if(epoch_num):
            self.history_loss[0].append(epoch_num)
            self.history_loss[1].append(cost)
            print("cost ",cost )
            print("dw1 ",dw1," \ndb1 ", db1," \ndw2 ", dw2," \ndb2 ", db2)
            print("{after} w1 ",self.w1," , b1 ",self.b1," , w2 ",self.w2," , b2 ",self.b2,)
        return cost
    
    def train_N(self, N = 1000, epoch_num = 100, epoch_print = 10):
        self.mk_train_dataset(N)
        self.show_dataset()
        for epoch in range(epoch_num):
#             self.mk_train_dataset(N)
            if (epoch % epoch_print == 0):
                self.train(epoch)
            else:
                self.train()
        return
    def show_loss(self):
        plt.plot(self.history_loss[0], self.history_loss[1], marker='.', c='red', label="Validation set Loss")
        return
    
    def show_dataset(self):
        idxs_1 = np.where(self.data_Y==1)
        idxs_0 = np.where(self.data_Y==0)
        X_0 = self.data_X[idxs_0]
        Y_0 = self.data_Y[idxs_0]

        X_1 = self.data_X[idxs_1]
        Y_1 = self.data_Y[idxs_1]

        plt.plot(X_0[:,0],X_0[:,1],"r^")
        plt.plot(X_1[:,0],X_1[:,1],"bx")
        plt.show()
        return
    
```

```python
XOR = ShallowNN(x_seeds,y_seeds,2,3,0.25)
XOR.train_N(1000,400,10)
XOR.show_loss()
```
